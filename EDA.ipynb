```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_renamed_rev = pd.read_csv("processed_data.csv")
df_renamed_rev
```

```python
df_renamed_rev.drop('Unnamed: 0',axis = 1, inplace = True)
df_renamed_rev
```

```python
df_renamed_rev['Number of ingredients'] = df_renamed_rev.apply(lambda x: x.sum(),axis = 1)
df_renamed_rev
```

```python
#ig, ax = plt.subplots(figsize=(10,10))
sns_plot = sns.distplot(df_renamed_rev["Number of ingredients"], bins = 90)
fig = sns_plot.get_figure()
fig.savefig("output.pdf")
```

```python
number  = df_renamed_rev['Number of ingredients']
np.mean(number), np.median(number)
```

```python
df_renamed_rev['Number of ingredients'].kurt()
```

```python
df_renamed_rev.drop(columns = ['Number of ingredients'],inplace= True)
```

```python
df_renamed_rev.loc['Row_sum'] = df_renamed_rev.apply(lambda x: x.sum())
df_renamed_rev
```

```python
sorted_df_renamed_rev = df_renamed_rev.T.sort_values(by = 'Row_sum',ascending = False).T
sorted_df_renamed_rev
```

```python
x_0 = [sorted_df_renamed_rev.columns.values[i] for i in range(50)]
```

```python
y_raw = sorted_df_renamed_rev.iloc[38472,:50]

y_0 = [y_raw[i] for i in range(len(y_raw))]
```

```python
fig, ax = plt.subplots(figsize=(30,19))

g_1 = sns.barplot(x = x_0,
            y = y_0,)
#sns.set(font_scale= 1.9 )
#ax.set_xlabel('X_label', fontsize=10)
ax.tick_params(axis='y',labelsize=25) 
ax.tick_params(axis='x',labelsize=25) 
plt.xticks(rotation=73)
plt.show()
fig2 = g_1.get_figure()
fig2.savefig("frequency.pdf")
```

```python
df_renamed_rev.drop(index = ['Row_sum'],inplace = True)
df_renamed_rev
```

```python
from sklearn.decomposition import PCA 
pca=PCA(n_components=2)
sorted_df_renamed_rev.drop(index = ['Row_sum'],inplace = True)
newData=pca.fit_transform(sorted_df_renamed_rev.T)
```

```python
df_50_data = pd.DataFrame(newData[:50].T)
df_50_data.columns  = sorted_df_renamed_rev.columns.values[:50]
df_50_data
```

```python
txt = df_50_data.columns.values
x = df_50_data.loc[0,:]
y = df_50_data.loc[1,:]
fig,ax = plt.subplots(figsize=(15,10))
ax.scatter(x,y,color = 'r')
for i in range(50):
    ax.annotate(txt[i],(x[i],y[i]))
ax.tick_params(axis='y',labelsize=20) 
ax.tick_params(axis='x',labelsize=20)
plt.savefig('pca.pdf')
```
