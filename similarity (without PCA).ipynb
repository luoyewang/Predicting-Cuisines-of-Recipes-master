```python
# 计算cosine similarity
def CS_f(U_i,U_j):
    CS_i_j = np.dot(U_i.T,U_j)/((math.sqrt(np.dot(U_i.T,U_i)))*((math.sqrt(np.dot(U_j.T,U_j)))))
    return CS_i_j                                                              
```

```python
# 计算 asymmetric cosine similarity
def ACS_f(U_i,U_j):
    ACS_i_j = np.dot(U_i.T,U_j)/((math.pow(np.dot(U_i.T,U_i),0.05))*(math.pow(np.dot(U_j.T,U_j),0.95)))
    return ACS_i_j
```

```python
# 计算Jaccard similarity
def JS_f(U_i,U_j):
    U = U_i + U_j
    N = 0
    for u in U:
        if u > 0:
            N = N + 1
    JS_i_j = np.dot(U_i.T,U_j)/N
    return JS_i_j
```

```python
# 计算PMI相似度
def PMI_f(U_i,U_j):
    p_i_j = np.dot(U_i.T,U_j)/len(U_i)
    p_i = np.dot(U_i.T,U_i)/len(U_i)
    p_j = np.dot(U_j.T,U_j)/len(U_j)
    if p_i_j == 0:
        PMI_i_j = 0
    else:
        PMI_i_j = math.log((p_i_j/(p_i*p_j)))
    return PMI_i_j
```

```python
#计算相似度矩阵
def SIM_matrix(R, F_index):
    SIM = np.zeros([R.shape[1],R.shape[1]])
    if F_index == 'CS':
        for i in range(SIM.shape[1]):
            for j in range(SIM.shape[1]):
                SIM[i,j] = CS_f(R.iloc[:,i],R.iloc[:,j])
    elif F_index == 'ACS':
        for i in range(SIM.shape[1]):
            for j in range(SIM.shape[1]):
                SIM[i,j] = ACS_f(R.iloc[:,i],R.iloc[:,j])
    elif F_index == 'JS':
        for i in range(SIM.shape[1]):
            for j in range(SIM.shape[1]):
                SIM[i,j] = JS_f(R.iloc[:,i],R.iloc[:,j])
    elif F_index == 'PMI':
        for i in range(SIM.shape[1]):
            for j in range(SIM.shape[1]):
                SIM[i,j] = PMI_f(R.iloc[:,i],R.iloc[:,j])
    else: 
        print('Wrong similarity function choice')
    return SIM
```

```python
def TOP_k_index(Sim_matrix, Row, K_ingredient):
    Sim_matrix_i = Sim_matrix[Row,:]
    Index = np.argsort(Sim_matrix_i)[::-1]
    Top_k = Index[:K_ingredient]
    Top_k = Top_k.tolist()
    
    if Row in Top_k:
        Top_k.remove(Row)
        Top_k.append(Index[K_ingredient])
        
    return Top_k
```

```python
# 获得要随机删除的成分的序列号
def Random_delete_index(V):
    Delete_index = []
    for i in range(len(V)):
        if V[i] == 1:
            Delete_index.append(i)
    return Delete_index[random.randint(0,len(Delete_index)-1)]
```

```python
#获得某一菜跟它最契合的前N个成分，随机删除成分的名字，随机删除成分的排序
def TOP_ingredient(Test_R_matrix, Recipe_index, N, K_ingredient, Sim_matrix, Sim_ingredient_list):
    P_recipe = []
    Top_N_ingredient_name = []
    All_ingredient_name = []
    Recipe_vec = Test_R_matrix.iloc[Recipe_index]#获得某一个菜的向量表示
    Delete_index = Random_delete_index(Recipe_vec) #获得要随机删除的菜原料的序号
    Delete_name = Test_R_matrix.columns.values[Delete_index]#获得要删除的菜原料的名字
    
    Recipe_vec[Delete_index] = 0
    
    for i in range(Test_R_matrix.shape[1]):
        Numerator = 0
        Denominator = 0
        for j in Sim_ingredient_list[i]:
            Numerator = Numerator + Recipe_vec[j] * Sim_matrix[i,j]
            Denominator = Denominator + Sim_matrix[i,j]
        P_recipe_i = Numerator / Denominator#获得测试的菜跟第i个原料的适合程度
        P_recipe.append(P_recipe_i)
        
    Ingredient_index = np.argsort(P_recipe)[::-1]
    Top_N_ingredient = Ingredient_index[:N]#获得相似度靠前的前n个原料
    
    
    for s in Ingredient_index:
        All_ingredient_name.append(Test_R_matrix.columns.values[s]) #按从适合程度高到低排序所有成分
        
    for n in Top_N_ingredient:
        Top_N_ingredient_name.append(Test_R_matrix.columns.values[n]) #获得这个菜前n个适合原料的序列号
        
    for l in range(Test_R_matrix.shape[1]):
        if All_ingredient_name[l] == Delete_name:
            Ingredient_rank = l
    Ingredient_rank = Ingredient_rank + 1
    
    return  Top_N_ingredient_name, Delete_name, Ingredient_rank
```

```python
#获得要测试的菜谱的序列号
def Test_index(Test_R_matrix,Number):
    Test_index = []
    for i in range(Number):
        Test_index.append(random.randint(0,Test_R_matrix.shape[0]))
    return Test_index
```

```python
#返回召回率，位次的平均值，位次的中位数
def Test_f(Train_R_matrix, Test_R_matrix, Test_index , N, K, Sim_function_choice):
    Rank = []
    N_ingredient_list = []
    
    True_label = 0
    Whole_label = len(Test_index)
    
    S = SIM_matrix(Train_R_matrix,Sim_function_choice) #得到相似矩阵（161*161）
    
    for i in range(Train_R_matrix.shape[1]):
        N_ingredient_list.append(TOP_k_index(S,i,K)) #得到所有原料成分的K个相似原料成分的序列
        
    for Recipe_i in Test_index:
        Predict_name, True_name, Predict_rank = TOP_ingredient(Test_R_matrix, Recipe_i, N, K, S, N_ingredient_list)
        Rank.append(Predict_rank)
        if True_name in Predict_name:
            True_label = True_label + 1
    
    Recall = True_label / Whole_label
    
    return Recall, np.mean(Rank), np.median(Rank)
```

# Test part

```python
test_index = Test_index(df_renamed_rev_test, 10)
test_index
```

```python
start = time.process_time()
recall_test, mean_rank_test, median_rank_test = Test_f(df_renamed_rev_train, df_renamed_rev_test, test_index, 10, 50, 'JS')
print('Recall rate is :%.2f%%'%(recall_test*100))
print('The mean of rank is :%.2f'%(mean_rank_test))
print('The median of rank is :%.2f'%(median_rank_test))
end = time.process_time()
print('Running time: %s Seconds'%(end-start))
```
