```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_renamed_rev = pd.read_csv("processed_data.csv")
df_renamed_rev
```

```python
df_renamed_rev.drop('Unnamed: 0',axis = 1, inplace = True)
df_renamed_rev
```

```python
df_renamed_rev['Col_sum'] = df_renamed_rev.apply(lambda x: x.sum(),axis = 1)
df_renamed_rev
```

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.distplot(df_renamed_rev["Col_sum"])
```

```python
df_renamed_rev.drop(columns = ['Col_sum'],inplace= True)
```

```python
df_renamed_rev.loc['Row_sum'] = df_renamed_rev.apply(lambda x: x.sum())
df_renamed_rev
```

```python
sorted_df_renamed_rev = df_renamed_rev.T.sort_values(by = 'Row_sum',ascending = False).T
sorted_df_renamed_rev
```

```python
fig, ax = plt.subplots(figsize=(30,20))
sns.barplot(x = sorted_df_renamed_rev.columns.values,
            y = sorted_df_renamed_rev.loc['Row_sum'],)
plt.xticks(rotation=90)
plt.show()
```

```python
df_renamed_rev.drop(index = ['Row_sum'],inplace = True)
df_renamed_rev
```

```python
from sklearn.decomposition import PCA 
pca=PCA(n_components=2)
sorted_df_renamed_rev.drop(index = ['Row_sum'],inplace = True)
newData=pca.fit_transform(sorted_df_renamed_rev.T)
```

```python
df_50_data = pd.DataFrame(newData[:50].T)
df_50_data.columns  = sorted_df_renamed_rev.columns.values[:50]
df_50_data
```

```python
txt = df_50_data.columns.values
x = df_50_data.loc[0,:]
y = df_50_data.loc[1,:]
fig,ax = plt.subplots(figsize=(15,15))
ax.scatter(x,y,color = 'r')
for i in range(50):
    ax.annotate(txt[i],(x[i],y[i]))
```
