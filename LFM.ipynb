```python
def random_recipe_index(Rating_matrix,Delete_number):
    random.seed(0)
    Length = Rating_matrix.shape[0]
    Delete_index = []
    for i in range(Delete_number):
        Temp = random.randint(0,Length-1)
        Delete_index.append(Temp)
    return Delete_index
```

```python
def random_ingredient_index(Recipe_vec):
    Ingredient_choice_index = []
        
    for i in range(161):    
        if (Recipe_vec[i] == 1):
            Ingredient_choice_index.append(i)
   
            
    if len(Ingredient_choice_index) == 1:
        index = 0
    else:
        index = np.random.randint(0,len(Ingredient_choice_index))
        
    return Ingredient_choice_index[index]
```

```python
Delete_recipe_index = random_recipe_index(R,3000)
Delete_ingredient = []
for i in Delete_recipe_index:
    Recipe_i = R[i]
    Recipe_i_ingredient_index = random_ingredient_index(Recipe_i)
    Delete_ingredient.append(Recipe_i_ingredient_index)
    R[i,Recipe_i_ingredient_index] = 0
```

```python
def tanh(x):

    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
from numpy.linalg import norm

def error(R,P,Q,lamda=0.02):
    ratings = R.data
    rows = R.row
    cols = R.col
    e = 0 
    for ui in range(len(ratings)):
        rui=ratings[ui]
        u = rows[ui]
        i = cols[ui]
        if rui!= 0:
            e= e + pow(rui-tanh(np.dot(P[u,:],Q[:,i])),2)+\
                lamda*(pow(norm(P[u,:]),2)+pow(norm(Q[:,i]),2))
    return e
```

```python
def SGD(R, K, gamma, lamda, steps):
    
    M,N = R.shape
    P = np.random.rand(M,K)
    Q = np.random.rand(K,N)
    
    rmse = np.sqrt(error(R,P,Q,lamda)/len(R.data))
    print("Initial RMSE: "+str(rmse))
    
    for step in range(steps):
        for ui in range(len(R.data)):
            rui = R.data[ui]
            u = R.row[ui]
            i = R.col[ui]
            if rui != 0:
                pre_ui = tanh(np.dot(P[u,:],Q[:,i]))
                eui=rui - pre_ui
                
                P[u,:]=P[u,:]+gamma*2*(eui*Q[:,i]*(1-pre_ui**2)-lamda*P[u,:])
                Q[:,i]=Q[:,i]+gamma*2*(eui*P[u,:]*(1-pre_ui**2)-lamda*Q[:,i])
                
        rmse = np.sqrt(error(R,P,Q,lamda)/len(R.data))
        if rmse<0.5:
            break
    print("Final RMSE: "+str(rmse))
    return P,Q
```

```python
start = time.process_time()
P_pre,Q_pre = SGD(R_new,K=20, gamma=0.0007,lamda=0.01, steps=50)
end = time.process_time()
print('Running time: %s Seconds'%(end-start))
```

```python
R_pre_raw = np.matmul(P_pre, Q_pre)
R_pre = np.zeros((R_pre_raw.shape[0],R_pre_raw.shape[0]))
for i in range(R_pre_raw.shape[0]):
    for j in range(R_pre_raw.shape[1]):
        R_pre[i,j] = tanh(R_pre_raw[i,j])
R_pre
```

```python
def test_f(R_prediction, Recipe_index,Ingredient_index, K):
    True_label = 0
    Confident_label = 0
    Whole_label = len(Recipe_index)
    Length = len(Recipe_index)
    Rank = []
    for i in range(Length):
        Recipe_ingredient =  R_prediction[Recipe_index[i],Ingredient_index[i]]
        Recipe_vec = R_prediction[Recipe_index[i]]
        if Recipe_ingredient >= 0:
            Confident_label = Confident_label + 1
            Top_index = np.argsort(Recipe_vec)[::-1]
            Top_k = Top_index[:K]
            if Ingredient_index[i] in Top_k:
                True_label = True_label + 1
            for j in range(Length):
                if Top_index[j] == Ingredient_index[i]:
                    Rank.append(j+1)
     
    recall = True_label/3000
    
    return recall, np.mean(Rank),np.median(Rank),Confident_label
```

```python
recall, mean, median,label = test_f(R_pre,Delete_recipe_index,Delete_ingredient,10)
```
